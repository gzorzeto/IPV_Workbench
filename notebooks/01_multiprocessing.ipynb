```python
import multiprocess as mp
import numpy as np
from time import sleep
```

```python
total_timesteps = len(np.arange(0,8760,1))
ncpu = 6


def create_timestep_chunks(total_timesteps, ncpu):
    incre = int(total_timesteps/ncpu)
    incre_range = range(0,total_timesteps,incre)
    incre_chunks = [np.arange(i,i+incre) for i in incre_range]

    return incre_chunks

create_timestep_chunks(total_timesteps, 6)
```

```python
def main_sim(hoy):
    results_hoy1 = {hoy: np.arange(1,10,1)}
    results_hoy2 = {hoy: np.arange(0,1,.1)}
    results_hoy3 = {hoy: np.arange(5,50,5)}
    return results_hoy1, results_hoy2, results_hoy3

def wrap_loop(hoy_chunk):
    A_list = {}
    B_list = {}
    C_list = {}
    for hoy in hoy_chunk:
        sleep(0.05)
        A,B,C = main_sim(hoy)
        A_list.update(A)
        B_list.update(B)
        C_list.update(C)
    return A_list,B_list,C_list
    
def mp_hoy():
    total_timesteps = len(np.arange(0,1000,1))
    chunks = create_timestep_chunks(total_timesteps, 6)
    pool = mp.Pool(6)
    result = pool.map(wrap_loop, chunks)
    pool.close()
    return result

    
```
